## **Passo a passo**
---

###Podemos iniciar definindo um dicionário, para nomear as unidades e dezenas, além de se atentar ao caso particular dos numeros 11 ao 19 que não seguem uma regra fixa. Faremos desta maneira:




```
dicionario_numerico = {
    'unidade':{
        'um':1,
        'dois':2,
        'três' : 3,
        'quatro':4,
        'cinco':5,
        'seis':6,
        'sete':7,
        'oito':8,
        'nove':9,
        'dez':10
    },
    'onzeAodezenove':{
        'onze':11,
        'doze':12,
        'treze':13,
        'quatorze':14,
        'quinze':15,
        'dezeseis':16,
        'dezesete':17,
        'dezoito':18,
        'dezenove':19,

    },
    'vinteanoventa':{
        'vinte':20,
        'trinta':30,
        'quarenta':40,
        'cinquenta':50,
        'sessenta':60,
        'setenta':70,
        'oitenta':80,
        'noventa':90,

    },
    'cemAteNovecentos':{                         #Implementamos esta função para ser usada no exercicio

        'cento':100,
        'duzentos':200,
        'trezentos':300,
        'quatrocentos':400,
        'quinhentos':500,
        'seiscentos':600,
        'setecentos':700,
        'oitocentos':800,
        'novecentos':900,


    },

}
```
### Após a definição do dicionário, prosseguiremos solicitando a entrada do número ao usuário da seguinte manteira:

`x = input('Digite um numero ')` 
*Aqui atribuímos a variável 'x' como uma entrada que o usuário digitará (através do teclado)*

###Em seguida copiamos valor de 'x' a uma variavel auxiliar 'digitos', e  feito isso atribuiremos a 'x' o valor do tamanho deste numero, através do método 
```.len```. Exemplo: 


```
digitos = x                                ## atribui o valor de x a digitos
if x.isnumeric():                          ## verifica se entrada é um número 
    x = len(x)                             ## atribuia a x o tamanho do vetor
```

###Agora definiremos a função para transformar o dado numero por extenso, será a ```def UnidadeDezena``` descrita a seguir


```
def UnidadeDezena():

    if x == 1:                                                  #Verifica se o usuario digitou um unico numero
        for k, v in dicionario_numerico['unidade'].items():     #Percorre o item Usando um laço de repetição do tipo for no dicionários referente a unidade
            if int(digitos) == v:                               #A cada loop verifica se o item em questão refere-se ao número correspondente 
                print('-' * 10)
                print(k)                                        #Mostra na tela este número
                print('-' * 10)

    if x == 2 and digitos[0] == '1' and digitos[1] == '0':      #Faz duas verificações, se o número tem 2 digitos e se ele corresponde ao número 10
        print('-' * 10)
        print(list(dicionario_numerico['unidade'])[9])          #Imprime o numero dez na tela.
        print('-' * 10)

    if x == 2 and digitos[0] == '1':                            #vefirica se o numero tem dois dígitos e se ele esta compreendido entre 10 e 20;
        for k, v in dicionario_numerico['onzeAodezenove'].items(): #Percorre o vetor com um laço de repetição "for" referente aos itens no dicionario da seção "onzeAodezenove"
            valor = str(v)                                      #Atribui a uma variavel "valor" o conteudo em forma de string de v (que é o numero da repetição corrente)
            if digitos[1] == valor[1]:                          #Verifica se o conteudo das unidades do digito é igual a que está em valor 
                print('-' * 10)
                print(k)                                        #entrenado nesse if mostra-se na tela o valor de k que será o item correspondente na lista "onzeAodezenove"
                print('-' * 10)



    if x == 2 and digitos[0] != '1':                                 #Verifica se o numero possue 2 digitos e se o valor da dezena é diferente de 1, caso atenda estas condições prosseguimos.
        for k, v in dicionario_numerico['vinteanoventa'].items():    #percorre os itens do dicionários 
            valor = str(v)                                           #Atribue a variavel "valor" o conteudo em forma de string do item no dicionário na sessão das dezenas 
            if digitos[0] == valor[0]:                               #encontrado o valor no dicionário
                saveK = k                                            #Salvamos esta informação em uma variável chamada "saveK" com o valor de K corrente (pois refere-se a dezena do numero desejado)
                for k, v in dicionario_numerico['unidade'].items():  #Agora repitimos o laço de repetição que usamos para descobrir a unidade
                    if digitos[1] == str(v):
                        print('-' * 10)
                        print(saveK, 'e', k)                         #E nesta parte mostramos na tela o numero da seguinta maneira  "Dezena 'e' unidade"
                        print('-' * 10)

    if x == 2 and digitos[1] == '0':                                 #Aqui verificamos se o numero possue 2 digitos e se o digito das unidades é 0
        for k, v in dicionario_numerico['vinteanoventa'].items():    #Percorremos com um laço de repetição for os itens da sessão "vinteAonoventa" do dicionário
            valor = str(v)                                           #Atribuímos a variavel valor o conteudo em forma de string de v (que é o id correspondente desta sessão do dicionário)
            if digitos[0] == valor[0]:                               #Fazemos uma verificação  se o digito da unidade é semelhante em conteudo ao primeiro indice do valor
                print('-' * 10)
                print(k)                                             #Mostramos na tela o valor de K que corresponde ao numero por extenso que estamos procurando
                print('-' * 10)
```

###Agora podemos contrUir um loop para ficar executando este programa por quanto tempo quisermos da seguinte maneira



```
while (x>100 and x<0):
    UnidadeDezena()                    #Laço de repetição while para ler numeros entre 0 e 100, digitar um numero fora desta faixa finaliza o programa.
   
```
